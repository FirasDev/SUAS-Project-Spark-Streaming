```python
part = spark.read.csv(
    "data2/part-00000-004d5866-f4df-4792-a026-40e18d73a802-c000.csv",
    header=True,
    inferSchema=True,
)
```

```python
part.groupBy("diagnosis").count().show()
```

```python
dataSchema = part.schema
```

```python
dataSchema
```

```python
streaming = (
    spark.readStream.schema(dataSchema)
    .option("maxFilesPerTrigger", 1)
    .csv("data2/")
)
```

```python
diag_count_op1 = streaming.groupBy("diagnosis").count().orderBy(F.desc("diagnosis"))
```

```python
activityQuery = (
    diag_count_op1.writeStream.queryName("diag_count_op1")
    .format("memory")
    .outputMode("complete")
    .start()
)

import time

for x in range(50):
    _df = spark.sql(
        "SELECT * FROM diag_count_op1"
    )
    if _df.count() > 0:
        _df.show(50)
    time.sleep(0.5)
```

```python
spark.streams.active[0].isActive
```

```python
diag_count_opi2 = streaming.groupBy("radius_mean").count().orderBy(F.desc("radius_mean"))
```

```python
activityQuery2 = (
    diag_count_opi2.writeStream.queryName("diag_count_opi2")
    .format("memory")
    .outputMode("complete")
    .start()
)

import time

for x in range(10):
    _df = spark.sql(
        "SELECT * from diag_count_opi2"
    )
    if _df.count() > 0:
        _df.show(10)
    time.sleep(0.5)
```
