```python
part = spark.read.csv(
    "data/part-00000-0020b336-7b04-45bd-bfac-9fa6c8377c33-c000.csv",
    header=True,
    inferSchema=True,
)
```

```python
part.groupBy("diagnosis").count().show()
```

```python
dataSchema = part.schema
```

```python
dataSchema
```

```python
streaming = (
    spark.readStream.schema(dataSchema)
    .option("maxFilesPerTrigger", 1)
    .csv("data/")
)
```

```python
diag_count = streaming.groupBy("diagnosis").count().orderBy(F.desc("diagnosis"))
```

```python
activityQuery = (
    diag_count.writeStream.queryName("diag_count")
    .format("memory")
    .outputMode("complete")
    .start()
)

# include this in production
#activityQuery.awaitTermination()

import time

for x in range(10):
    _df = spark.sql(
        "SELECT * FROM diag_count"
    )
    if _df.count() > 0:
        _df.show(10)
    time.sleep(0.5)
```
